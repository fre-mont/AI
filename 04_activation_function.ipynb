```python
import numpy as np
import matplotlib.pyplot as plt
```

### 계단함수


```python
# def step(x):
#     if x > 0.00001: return 1
#     else : return 0
    
def step(x):
    result = x > 0.000001    # True(1) or False(0)
    return result.astype(np.int)  # 정수 타입으로 반환 
```

```python
x = np.arange(-10.0, 10.0, 0.1)
y = step(x)
plt.plot(x,y)
plt.show()
```

### 시그모이드 함수


```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

### tanh 함수
- 넘파이에서 제공


```python
x = np.linspace(-np.pi, np.pi, 60)   # 탄젠트 함수의 주기는 -pi ~ pi
y = np.tanh(x)
plt.plot(x, y)
plt.show()
```

### ReLU 

```python
def relu(x):
    return np.maximum(x, 0)

x = np.arange(-10.0, 10.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```
